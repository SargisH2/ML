```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("heart.csv")
data
```

```python
data['Sex'] = data['Sex'].apply(lambda x: int(x=='M')).astype('int8', errors='raise')
```

```python
c_p_t = np.unique(data['ChestPainType'])
for i in c_p_t:
    data[i] = data['ChestPainType'].apply(lambda x: int(x == i)).astype('int8')

data
```

```python
# pd.concat([data, pd.get_dummies('ChestPainType', dtype='int8')], axis=0)
```

```python
data.drop('ChestPainType', axis=1, inplace = True)
```

```python
# data.describe()
data
```

```python
data.isnull().sum()
```

```python
# print(np.unique(data['RestingECG']))

data = pd.concat([data, pd.get_dummies(data['RestingECG'], dtype='int8')], axis=1)
```

```python
data.drop('RestingECG', axis=1, inplace=True)
data
```

```python
data['ExerciseAngina'] = data['ExerciseAngina'].apply(lambda x: int(x=='Y')).astype('int8', errors='raise')
data['ST_Slope'] = data['ST_Slope'].apply(lambda x: int(x=='Up')).astype('int8', errors='raise')
```

```python
data['RestingBP'] = data['RestingBP'] / np.max(data['RestingBP'])
data['Cholesterol'] = data['Cholesterol'] / np.max(data['Cholesterol'])
data['MaxHR'] = data['MaxHR'] / np.max(data['MaxHR'])
data['Age'] = data['Age'] / np.mean(data['Age'])
```

```python
data
```

```python
result = data['HeartDisease']
data.drop('HeartDisease', axis = 1, inplace = True)
data = pd.concat([data, result], axis=1)
```

```python
data.to_csv('clean_data.csv', index = False)
```
