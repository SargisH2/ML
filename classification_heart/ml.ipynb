```python
import pandas as pd

from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier

from sklearn.metrics import accuracy_score, f1_score
```

```python
data = pd.read_csv('clean_data.csv')
data
```

```python
x, y = data.loc[:,:"ST"].values, data['HeartDisease'].values
x_train, x_test, y_train, y_test = train_test_split(x, y)
x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

```python
model_knn = KNeighborsClassifier(n_neighbors=5)
model_log = LogisticRegression()
model_tree = DecisionTreeClassifier(max_depth=4)
model_forest = RandomForestClassifier(max_depth=3, n_estimators=150)
```

```python
model_knn.fit(x_train, y_train)
print(accuracy_score(y_train,model_knn.predict(x_train)), accuracy_score(y_test,model_knn.predict(x_test)))
print(f1_score(y_train,model_knn.predict(x_train)), f1_score(y_test,model_knn.predict(x_test)))
```

```python
model_log.fit(x_train, y_train)
print(accuracy_score(y_train,model_log.predict(x_train)), accuracy_score(y_test,model_log.predict(x_test)))
print(f1_score(y_train,model_log.predict(x_train)), f1_score(y_test,model_log.predict(x_test)))
```

```python
model_tree.fit(x_train, y_train)
print(accuracy_score(y_train,model_tree.predict(x_train)), accuracy_score(y_test,model_tree.predict(x_test)))
print(f1_score(y_train,model_tree.predict(x_train)), f1_score(y_test,model_tree.predict(x_test)))
```

```python
model_forest.fit(x_train, y_train)
print(accuracy_score(y_train,model_forest.predict(x_train)), accuracy_score(y_test,model_forest.predict(x_test)))
print(f1_score(y_train,model_forest.predict(x_train)), f1_score(y_test,model_forest.predict(x_test)))
```
